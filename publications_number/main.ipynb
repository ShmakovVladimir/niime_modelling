```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
queries = ['memristor.csv', 
           'memristor modelling.csv', 
           'memristor modelling using deep learning algorithms.csv',
           'memristor compact model.csv']


fig, ax = plt.subplots(2, 2, figsize = (12, 12), dpi = 200, sharex=True)

for table_name, plotable in zip(queries, ax.ravel()):
    data = pd.read_csv(table_name)
    data = data[data.year < 2024]
    plotable.scatter(data.year, data.publications, s = 100, edgecolors='black')
    plotable.grid()
    plotable.set_xticks(np.arange(2010, 2024, 2))
    plotable.set_title(table_name[:-4], fontsize = 16)

fig.tight_layout()
```
